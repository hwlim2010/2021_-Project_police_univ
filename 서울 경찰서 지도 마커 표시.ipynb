```python
df = pd.read_csv('경찰청_경찰관서 위치, 주소_20200409.csv', encoding='EUC-KR')
df
```

```python
df_seoul = df[df['청'] == '서울청']
df_seoul
```

```python
df_seoul.index = df_seoul.지구대파출소
df_seoul
```

```python
df_seoul.index
```

```python
seoul_map = folium.Map(location=[37.55,126.98], tiles='Stamen Terrain', zoom_start=12)
```

```python
for name, lat, lng in zip(df_seoul.index, df_seoul.Y좌표, df_seoul.X좌표):
    folium.Marker([lat, lng], popup=name).add_to(seoul_map)
```

```python
seoul_map.save('/seoul_poilcemap1.html')
```
